```python
data_no_loc_SA = dfs_all.drop(data_loc_SA.index)
new_lat = data_no_loc_SA.lng.tolist()
new_lng = data_no_loc_SA.lat.tolist()
```

```python
data_no_loc_SA.lng
```

```python
data_no_loc_SA['lat'] = new_lat
data_no_loc_SA['lng'] = new_lng
```

```python
data_no_loc_SA.lng
```

```python
frames = [data_loc_SA, data_no_loc_SA]
df_result = pd.concat(frames)
df_result
```

```python
# data_no_loc_SA = dfs_all.drop(data_loc_SA.index)
# data_no_loc_SA.head()
```

```python
df_all_loc = df_result.copy()
data_new_loc_SA = df_all_loc[ df_all_loc.apply(lambda row: (16<=row['lat']<= 17.5) and (41<=row['lng']<=42.3), axis=1) ]
data_new_no_loc_SA = df_all_loc.drop(data_new_loc_SA.index)

data_loc_seim_final_SA = data_new_no_loc_SA[ data_new_no_loc_SA.apply(lambda row: (17.8<=row['lat']<= 18) and (41<=row['lng']<=42), axis=1) ]
data_loc_final_SA = data_new_no_loc_SA.drop(data_loc_seim_final_SA.index)
```
